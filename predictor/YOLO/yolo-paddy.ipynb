Final train using new model and best hyperparameters

```python
import yaml

with open("/kaggle/working/runs/segment/tune/best_hyperparameters.yaml") as f:
    best_hyp = yaml.safe_load(f)

model = YOLO("yolo11x-seg.pt")

# Train with tuned hyperparameters
results = model.train(
    data="/kaggle/input/diseased-leaves/data.yaml",
    epochs=300,
    patience=50,
    imgsz=640,
    batch=16,
    workers=2,
    device="0,1",
    save=True,
    save_period=-1,
    verbose=True,
    project="/kaggle/working/",
    name="last_train_dataset_paddy_leaves_2",  
    **best_hyp
)
```
